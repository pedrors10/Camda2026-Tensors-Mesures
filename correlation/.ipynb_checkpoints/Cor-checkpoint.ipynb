```python
classes = sorted(meta["disease"].unique())

print(f"{len(classes)} clases encontradas")
```

```python
tax_submatrices = {}
path_submatrices = {}

for cls in classes:

    samples = meta.loc[
        meta["disease"] == cls,
        "sample_id"
    ].tolist()

    tax_cols = [
        tax.columns[0],
        *[s for s in samples if s in tax.columns]
    ]

    path_cols = [
        path.columns[0],
        *[s for s in samples if s in path.columns]
    ]

    tax_submatrices[cls] = tax[tax_cols]

    path_submatrices[cls] = path[path_cols]
```

```python
tax_corr = {}
tax_adj = {}

threshold = 0.7

for cls, df in tax_submatrices.items():

    X = df.iloc[:,1:]

    corr = X.T.corr(method="pearson")

    adj = (
        np.abs(corr) >= threshold
    ).astype(int)

    np.fill_diagonal(adj.values, 0)

    names = df.iloc[:,0]

    corr.index = names
    corr.columns = names

    adj.index = names
    adj.columns = names

    tax_corr[cls] = corr
    tax_adj[cls] = adj
```

```python
path_corr = {}
path_adj = {}

threshold = 0.7

for cls, df in path_submatrices.items():

    X = df.iloc[:,1:]

    corr = X.T.corr(method="pearson")

    adj = (
        np.abs(corr) >= threshold
    ).astype(int)

    np.fill_diagonal(adj.values, 0)

    names = df.iloc[:,0]

    corr.index = names
    corr.columns = names

    adj.index = names
    adj.columns = names

    path_corr[cls] = corr
    path_adj[cls] = adj
```

```python
adj = tax_adj["healthy"]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.from_pandas_adjacency(adj)

plt.figure(figsize=(12,12))

pos = nx.spring_layout(
    G,
    seed=42
)

nx.draw_networkx_nodes(
    G,
    pos,
    node_size=20
)

nx.draw_networkx_edges(
    G,
    pos,
    alpha=0.3
)

plt.title("Taxon Co-occurrence Network - Healthy")
plt.axis("off")
plt.show()
```

```python
import numpy as np
import networkx as nx

degree_histograms = {}

for cls, adj in tax_adj.items():

    G = nx.from_pandas_adjacency(adj)

    degrees = np.array([d for _, d in G.degree()])

    degree_histograms[cls] = degrees
```

```python
max_degree = max(
    deg.max()
    for deg in degree_histograms.values()
)

bins = np.arange(max_degree + 2)
```

```python
histograms = {}

for cls, degrees in degree_histograms.items():

    hist, _ = np.histogram(
        degrees,
        bins=bins
    )

    histograms[cls] = hist
```

```python
def chi2_distance(p, q):

    mask = (p + q) > 0

    return np.sum(
        (p[mask] - q[mask])**2 /
        (p[mask] + q[mask])
    )
```

```python
import pandas as pd

classes = sorted(histograms.keys())

D = pd.DataFrame(
    index=classes,
    columns=classes,
    dtype=float
)

for c1 in classes:
    for c2 in classes:

        D.loc[c1, c2] = chi2_distance(
            histograms[c1],
            histograms[c2]
        )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))

sns.heatmap(
    D,
    cmap="plasma",
    annot=False,
    fmt=".1f"
)

plt.title(
    "Chi-square Distance Between Degree Distributions"
)

plt.tight_layout()

plt.savefig(
    "disease_correlation_heatmap.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```

```python
pure_classes = [
    c for c in D.index
    if ";" not in c
]

D_pure = D.loc[pure_classes, pure_classes]

print("Número de clases puras:", len(pure_classes))
D_pure
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

sns.heatmap(
    D_pure,
    cmap="plasma",
    annot=False,
    fmt=".2f"
)

plt.title("Chi-square Distance Between Pure Classes")

plt.tight_layout()
plt.savefig(
    "Chi-squareDistanceBetweenPureClasses.png",
    dpi=300,
    bbox_inches="tight"
)
```
